```python
import pandas as pd
df = r"C:\Users\RSC\Desktop\data sheets\refernce sheets\python with sql and excel ttask\Olympics_Data_1600.xlsx"
df=pd.read_excel(df)
df
```

```python
#data cleaning
df.info()
df.isnull().sum()
df.duplicated()
```

```python
#data cleaning
df1 = df.drop_duplicates()
df1
```

```python
df1.isnull().sum()
```

```python
df1 = df1.dropna(subset=["Country", "Sport", "Athletes", "Gender"])
print(df1.isnull().sum())
```

```python
1. Histogram
2. Box Plot
3. Violin Plot
4. Scatter Plot
5. Correlation Heatmap
6. Bar Chart
7. Line Chart
8. Pie Chart or Treemap
```

```python
#histgrom
import matplotlib.pyplot as py
import seaborn as sns
py.hist(df1["Sport"])
py.title("histogram of sports")
py.show()
```

```python
import matplotlib.pyplot as py
import seaborn as sns
sns.violinplot(y=df1["Athletes"])
py.title("Details of Athletes")
py.show()

```

```python
import matplotlib.pyplot as py
py.scatter(x="Gold",y="Sport",data=df1)
py.title("Gold Medalist in every Sports")
py.show()
```

```python
import matplotlib.pyplot as py
import seaborn as sns
sns.boxplot(x="Country",y="Sport",data=df1)
py.title("sports held in Countries")
py.show()
```

```python
import matplotlib.pyplot as py
import seaborn as sns
sns.heatmap(df1[["Athletes","Gold","Bronze","Silver"]].corr(),annot =True)
py.title("heatmap")
py.show()

```

```python
import matplotlib.pyplot as py
import seaborn as sns
gold_medalists = df1.groupby("Country")["Gold"].sum()
top10 = gold_medalists.sort_values(ascending=False).head(10)
top10.plot(kind="bar")
py.title("Top 10 Gold Medal Countries")
py.xlabel("Country")
py.ylabel("Gold Medals")
py.show()
```

```python
import matplotlib.pyplot as py
import seaborn as sns
sns.lineplot(x="Gold",y="Bronze",data=df1)
py.title("NO of Gold andd Bronze Medalist")
py.show()
```

```python
import matplotlib.pyplot as py
import seaborn as sns
medals =[df1["Gold"].sum(),df1["Silver"].sum(),df1["Bronze"].sum()]
labels =["Gold", "Silver", "Bronze"]
py.pie(medals,labels=labels)
py.title("No.of Medalist in all the Sports")
py.show()
```
